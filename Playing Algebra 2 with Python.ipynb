#### Vertex Form
    If we know the vertex is (h,k), we may write:
$$y = a(x-h)^2 + k$$

    The vertex is (h,k), the axis of symmetry is: x = h

#### Intercept Form
    If we know interception points (a,0) and (b,0), we may write:
$$y = c(x-a)(x-b)$$

#### Factoring:
    Factoring is to find intercept form. (x-k) is a factor.

$$ y = (x+a)(x+b) $$

$$ y = x^2 + (a+b)x + ab $$

For example $ y = x^2 + 7x + 12 = (x+3)(x+4)$

#### 1.8 Use the Quadratic Formula

Let a,b,c are real number, such that a≠0. The quadratic function is $y = ax^2 + bx + c $

To solve the function, we let y = 0, that is $ax^2 + bx + c = 0 $

The we may have $ x = \frac{-b ± \sqrt{b^2-4ac} }{2a}$

Here, we say D, the "Discriminant" for quadratic function, is $b^2-4ac$ 
- If D > 0 : Two real root
- If D = 0 : Same root
- If D < 0 : No real root

##### Addition Material: What "No real root" really means...

We already have a quadratic function $y = ax^2 + bx + c$

Now we rewrite it as: $y = aw^2 + bw + c$, where $w = x + jz$

$$y = a(x + jz)^2 + b(x + jz) + c = a(x^2 - z^2 +2jxz) + b(x + jz) + c$$

$$y = ax^2 + bx + (c-az^2) + j(2xy+bz)$$

Good! We know that z = 0 is just a special case.
If z = 0, $y = ax^2 + bx + c$ . 


Let's check an example: $a = 1, b = 0, c = 4$

As $(b^2-4ac) < 0 \to $ There is no "Real Root"

Let's play with complex number:

$y = (x+jz)^2 + 4$

$y = x^2 - z^2 + 4 + jxz$

Root means y = 0, that is 0 in real part, and 0 in imaginary part (y = 0 + 0j)

Imaginary part: $xz = 0$

Real part: $(x+jz)^2 + 4 = x^2 - z^2 + 4 = 0$

We may infer:
1. x = 0 or z = 0 (Come from xz=0)
2. When x = 0, then z = ±2; When z = 0, then x = 2j, which is not "real", so z≠0

So, we have roots: 2j and -2j

### Ch3. Rational Exponents and Radical Functions

#### Rational Exponents 

Addition $ a^{m+n} = a^{m} \cdot a^{n}$

Subtraction $ a^{m-n} = \frac{a^{m}}{a^{n}}, a≠0$

Multiplication $(a^{m})^n = a^{m \cdot n}$

Division $a^{1/n} = \sqrt[n]{a} $

$a^{m/n} = (a^{1/n})^{m} = (\sqrt[n]{a})^{m} $

$a^{-m/n} = \frac{1}{(a^{1/n})^{m}} = \frac{1}{(\sqrt[n]{a})^{m}} , a≠0$


#### Operation of function

Addition $h(x) = f(x) + g(x) $

Subtraction $h(x) = f(x) - g(x) $

Multiplication $h(x) = f(x) \cdot g(x) $

Division $h(x) = \frac{f(x)}{g(x)} $

Composition $h(x) = g(f(x))$

** What is Composition g(f(x)) means? **

Feed the output of f(x) to g(x). Cascaded the two function: $ x \to f() \to g() \to y $

If $f(x) = x-3$, $g(x) = x^2 + 2$ then

$ h(x) = g(f(x)) = (f(x))^2+2 = (x-3)^2+2 = x^2 - 6x + 11 $

#### Inverse Function

A function describes the relationship of input and output. Inverse function means "given the output, you may get the input".

$f^{-1}(x)$ is the inverse function of $f(x)$


##### How to get $f^{-1}$ ?

    Step 1: replace y by x, and replace x by y
    Step 2: re-arrange it as y = ...


##### Check the composition of inverse function is x: $ f(f^{-1}(x)) = x $

For example, $y = f(x) = x^3$

Step 1: $x = y^3$

Step 2: $y = x^{1/3}$, So the $f^{-1}(x) = \sqrt[3]{x}$




```python
################################################################################
# Let's see a Inverse Function
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.18
################################################################################
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0, 10., 0.1)

f = (x**3)/10

# g is the inverse function of f
# g = (10x)^(1/3), use abs is to avoid error when x < 0
g = ((10*abs(x))**(1/3)) * (x/abs(x)) 

# Plot to figure
plt.figure(figsize=(8, 6), dpi=80)
plt.title('Inverse function')
plt.plot(x, f, 'b',label='$y=x^{3}/10$')
plt.plot(x, g, 'g',label='$y=(10^{1/3})x^{1/3}$')
plt.plot(x, x, 'r--',label='$y=x$',alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-10,10]) # Limit the display range
plt.ylim([-10,10]) # Limit the display range
plt.legend(loc=2)
plt.grid(True)
plt.show()
```

**Binominal theorem (二項式定理)**:

${\displaystyle (x+y)^{n}={n \choose 0}x^{n}y^{0}+{n \choose 1}x^{n-1}y^{1}+{n \choose 2}x^{n-2}y^{2}+\cdots +{n \choose n-1}x^{1}y^{n-1}+{n \choose n}x^{0}y^{n},} $

${\displaystyle {\tbinom {n}{k}} = {\frac  {n!}{k!(n-k)!}} = _nC_k}$

Use Binominal theorem in e: ($y=1,x={\frac{1}{n}}$)

${\displaystyle (1+{\frac{1}{n}})^{n}={n \choose 0}{\frac{1}{n}}^{0}+{n \choose 1}{\frac{1}{n}}^{1}+{n \choose 2}{\frac{1}{n}}^{2}+\cdots +{n \choose {n-1}}{\frac{1}{n}}^{n-1}+{n \choose n}{\frac{1}{n}}^{n},} $

${\displaystyle (1+{\frac{1}{n}})^{n} = \frac{1}{0!}+\frac{1}{1!}+\frac{1}{2!}+\cdots+\frac{1}{(n-1)!}+\frac{1}{n!}}$


$e = \sum_{n=0}^{\infty}\frac{1}{n!}$

#### Logarithmic Functions

If $f(x) = b^x$, the inverse function $f^{-1}(x)=log_b(x)$, logarithm of x with base b, or we say "log base b of x"

For example,

$ 2^3 = 8 \to log_2(8) = 3$

##### Logarithm properties
- Product property: $log_b(mn) = log_b(m) + log_b(n)$
- Quotient property: $log_b(\frac{m}{n}) = log_b(m) - log_b(n)$
- Power property: $log_b(m^n) = n \cdot log_b(m)$

##### $log_b(0) \to -\infty$ 

$log_b(0)$ is not defined. Because there is no y meets $b^y = 0$

But, if we say $b^y \to 0$, then, $log_b(0) \to -\infty$ 

##### $log_b(1) = 0$ 

Because $b^0 = 1$, logarithm function always pass point (1,0)


```python
################################################################################
# Let's see a Logarithmic Function
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.18
################################################################################
import numpy as np
import matplotlib.pyplot as plt

step = 0.1 # give smaller step, one x will close to 0, log(0) -> -inf

x = np.arange(0.1, 4., step)

p = 0*x
q = np.log(x)    # Natural log
r = np.log10(x)  # 10-based log 
s = np.log2(x)   # 2-based log

# Plot to figure
plt.figure(figsize=(8, 6), dpi=80)
plt.title('Logarithmic function')
plt.plot(x, p, 'b',alpha=0.5)
plt.plot(x, s, 'm',label='$y=log_{2}(x)$')
plt.plot(x, q, 'g',label='$y=ln(x)$')
plt.plot(x, r, 'r',label='$y=log_{10}(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=4)
plt.grid(True)
plt.show()
```

##### Arithmetic sequence 等差數列 : the distance between 2 terms is the same

$a_n = a_1 + (n-1)\cdot r$

Sum of arithmetic sequence: Area of trapezoid (梯形面積公式)

$ S_n = n \cdot (\frac{a_1+a_n}{2})$


##### Geometric sequence 等比數列：the ratio between 2 terms is the same

$a_n = a_1r^{n-1}$

Sum of geometric sequence:

$ S_n = a_1(\frac{1-r^n}{1-r})$

If r < 1, $n \to \infty$ , $ S_n = a_1(\frac{1}{1-r})$ 


**Wait a minute, why the name "Arithmetic" and "Geometric"?**

In a arithmetric sequence, each number is the arithmetic mean of the number adjacent to it. For example, in the arithematic sequence "1,3,5,7,9... ", $5 = \frac{3+7}{2}$ (arithematic mean)


In a geometric sequence, each number is the geometric mean of the numbers adjacent to it. For example, in the geometric sequence "1,2,4,8,16... ", $4 = \sqrt{2 \times 8}$ (geometric mean)




**I know arithmetic mean(算數平均),but what is geometric mean(幾何平均)?**

Geometric mean is to find a square that have same area as a rectangle. $d^2 = a \times b$

$Mean_{arithmetic} = \frac{\sum_{i=1}^{n}x_i}{n} $

$Mean_{geometric} = (\prod_{i=1}^{n}x_i)^{\frac{1}{n}} $


```python
################################################################################
# Geometric series
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1., 50., 1)
r=1.1
y_11 = (r**x)
y_12 = 2 * (r**x)

r = 1.1
r2 = 1.2
y_21 = (r**x)
y_22 = ((r2)**x) 

a_1 = 1.1
r = 1.2
y_31 = a_1 * (r**x)
y_32 = a_1 * ((-r)**x) 

a_1 = 1
r = 0.9
y_41 = a_1 * (r**x) 
y_42 = a_1 * ((-r)**x) 

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title('Geometric Series')
plt.subplot(411)
plt.plot(x,y_11,'go',x,y_12,'r^')
plt.grid(True)
plt.ylabel('a')


plt.subplot(412)
plt.plot(x,y_21,'go',x,y_22,'r^')
plt.grid(True)
plt.ylabel('r')

plt.subplot(413)
plt.plot(x,y_31,'go',x,y_32,'r^--')
plt.grid(True)
plt.ylabel('|r| > 1')

plt.subplot(414)
plt.plot(x,y_41,'go',x,y_42,'r^--')
plt.grid(True)
plt.ylabel('|r| < 1')

plt.xlabel('x')
plt.tight_layout()
plt.show()
```

#### Find Sums of Infinite Geometric Series

Why don't we find the sum of infinite arithematic series? If $n \to \infty$, $ S_n = n \cdot (\frac{a_1+a_n}{2})$, $S_{\infty} \to \infty$

Then, sum of infinite geometric series wont' go infinite?

$a_n = a_1r^{n-1}$

$ S_n = a_1(\frac{1-r^n}{1-r})$

If r = 0, $ S_{\infty} = a_1(\frac{1-0^{\infty}}{1-0}) = a_1$

If |r| < 1, $ S_{\infty} = a_1(\frac{1}{1-r}) $

If |r| ≥ 1, the series has no **definite** sum


```python
################################################################################
# Infinite Geometric series
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import numpy as np
import matplotlib.pyplot as plt

# r=0.9
r=0.8
x = np.linspace(0, 20, 20)
y = (r**x)

# for plotting
x=np.append(x,x[-1])
y=np.append(y,0)
x=np.append(x,0)
y=np.append(y,0)

plt.figure(figsize=(8, 6), dpi=80)
plt.bar(x, y, width=1.0, bottom=None,color='g',alpha=0.25)
plt.fill(x, y, zorder=10,color='g',alpha=0.75)
plt.grid(True, zorder=5)
plt.title('Infinite Geometric Series')
plt.ylabel('y')
plt.xlabel('x')
plt.text(20,0.01,'Small enough?',color='r')
plt.show()
```

#### Parabolas (拋物線)
A parabola is defined by a point (focus point) and a line (directrix). Any point on a parabola has same distance to focus point and to directrix.


```python
################################################################################
# Parabola
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-4., 4.1, 0.1)
y = 0.25*x**2
directrix = 0*x + (-1)

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.gca()
ax.cla()

# Focus
plt.plot([0],[1],'ro')
plt.text(-0.5,1.2,'Focus',color='r')

# Directrix
plt.plot(x, directrix, color='r',alpha=0.75)
plt.text(-3.9,-.9,'Directrix',color='r')

# Parabola
plt.plot(x, y,color='g',alpha=0.75)

# Circle 1
px = 1
py = 0.25*(px**2)
plt.plot([px,px,0],[-1,py,1],'ro-')
circle1 = plt.Circle(xy=(px, py),radius=(py+1), color='m',fill=False,alpha=0.5)
ax.add_artist(circle1)

# Circle 2
px = 2.5
py = 0.25*(px**2)
plt.plot([px,px,0],[-1,py,1],'co-')
circle2 = plt.Circle(xy=(px, py),radius=(py+1), color='c',fill=False,alpha=0.5)
ax.add_artist(circle2)

plt.grid(True)
plt.title('Parabolas')
plt.ylabel('y')
plt.xlabel('x')
plt.axis('equal')
plt.show()
```

#### Ellipses

Ellipse is "Two focus" cirle, say $f_1,f_2$

Ellipse is the set of point P in the plane that $d_1 + d_2 = foci = constant$

where $d_1,d_2$ are distances from P to $f_1$ and P to $f_2$

Ellipse is $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

Center of ellipse c:

If a > b: $a^2 - b^2 = c^2 $


```python
################################################################################
# Ellipse
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import numpy as np
import matplotlib.pyplot as plt

a,b = 5.,3.
c = np.sqrt(a**2 - b**2)

x = np.arange(-a, a, 0.1)
y = pow((1.-(x**2)/(a**2)),0.5)*b

# Append (a,0) because np.sqrt(0) is not avaliable
x = np.append(x,a)
y = np.append(y,0.0)

# Mirror to negative
y_min = -y

# Pick two Test Point:
tp_x = [x[50], x[80]]
tp_y = [y[50], y[80]]

d1 = np.sqrt((-c-tp_x[0])**2 + (tp_y[0])**2)
d2 = np.sqrt((c-tp_x[0])**2 + (tp_y[0])**2)

d3 = np.sqrt((-c-tp_x[1])**2 + (tp_y[1])**2)
d4 = np.sqrt((c-tp_x[1])**2 + (tp_y[1])**2)

print('d1+d2 =',d1+d2)
print('d3+d4 =',d3+d4)

plt.figure(figsize=(6, 6), dpi=80)

plt.plot(x,y,'g')     # Draw Ellipse upper part
plt.plot(x,y_min,'g') # Draw Ellipse lower part
plt.plot([-a,a],[0,0],'r') # Draw Center line

plt.plot([-c,tp_x[0],c],[0,tp_y[0],0],'co-') # Draw Test Point 0
plt.plot([-c,tp_x[1],c],[0,tp_y[1],0],'mo-') # Draw Test Point 1
plt.text((tp_x[0]-c)/2,0.2+tp_y[0]/2,'d1',color='c')
plt.text((tp_x[0]+c)/2,tp_y[0]/2,'d2',color='c')
plt.text((tp_x[1]-c)/2,0.2+tp_y[1]/2,'d3',color='m')
plt.text((tp_x[1]+c)/2,tp_y[1]/2,'d4',color='m')

plt.grid(True)
plt.title('Ellipse')
plt.ylabel('y')
plt.xlabel('x')
plt.axis('equal')
plt.show()
```

### Ch9. Trigonometric Ratios and Functions

#### Law of Sines

The area T of any triangle can be written as one half of its base times its height. Selecting one side of the triangle as the base, the height of the triangle relative to that base is computed as the length of another side times the sine of the angle between the chosen side and the base. Thus, depending on the selection of the base the **area of the triangle** can be written as any of:

${\displaystyle T={\frac {1}{2}}b(c\sin A)={\frac {1}{2}}c(a\sin B)={\frac {1}{2}}a(b\sin C)\,.}$


Multiplying these by 2/abc gives

${\displaystyle {\frac {2T}{abc}}={\frac {\sin A}{a}}={\frac {\sin B}{b}}={\frac {\sin C}{c}}\,.} $

#### Heron's Area Formula

Heron's formula states that the area of a triangle whose sides have lengths a, b, and c is

${\displaystyle A={\sqrt {s(s-a)(s-b)(s-c)}},},$

where s is the **semiperimeter** of the triangle; that is,

${\displaystyle s={\frac {a+b+c}{2}}.}$


**semiperimeter**

The area A of any triangle is the product of its **inradius** (the radius of its inscribed circle) and its semiperimeter:

${\displaystyle A=rs.}$

#### Inverse Function

Trigonometric functions **cannot** have inverse function. It is many to one mapping. But if we limit x in the range the inverse functions exist.

For sine function sin(x), x ranges $(-\frac{\pi}{2},\frac{\pi}{2})$, the inverse function of sine is call **arcsine**, note as arcsin() or $sin^{-1}()$

For cosine function cos(x), x ranges $(0,{\pi})$,the inverse function of cosine is call **arccosine**, note as arccos() or $cos^{-1}()$


#### Sinusoid
Graphs of sine and cosine function are called sinusoid.

$ y = a \cdot sin(b(x-h))+k$ or $ y = a \cdot cos(b(x-h))+k$

|a| is amplitude

$\frac{2\pi}{b}$ is period, or $\frac{b}{2\pi}$ is grequency, (b > 0)

h is horizontal shift

k is vertical shift


```python
################################################################################
# Sinusoid
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
import numpy as np

fs = 100 # sample rate: How many point in 1 cycle 
f = 4 # the frequency of the signal (Hz)

a = 3.0
b = 2*np.pi*f
h = np.pi/3
k = 1.0

x = np.arange(fs)
rad = [b * (i/fs)-h for i in np.arange(fs)]
tick = [2 * (i/fs) for i in np.arange(fs)] # tick is only for drawing x axis

y = a * np.sin(rad)+k

plt.figure(figsize=(8, 6), dpi=80)
ax = plt.gca()
ax.cla()
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=0.5))

plt.title('Sinusoid function')
plt.plot(tick, y, 'g-',label='Sinusoid')
plt.grid(True)
plt.show()
```

```python
import pyaudio
import numpy as np

p = pyaudio.PyAudio()

volume = 0.5     # range [0.0, 1.0]
fs = 44100       # sampling rate, Hz, must be integer
duration = 1.0   # in seconds, may be float
f = 440.0        # sine frequency, Hz, may be float

# generate samples, note conversion to float32 array
samples = (np.sin(2*np.pi*np.arange(fs*duration)*f/fs)).astype(np.float32)

# for paFloat32 sample values must be in range [-1.0, 1.0]
stream = p.open(format=pyaudio.paFloat32,
                channels=1,
                rate=fs,
                output=True)

# play. May repeat with different volume values (if done interactively) 
stream.write(volume*samples)

stream.stop_stream()
stream.close()

p.terminate()
```

#### Euler Identity $e^{i\pi }+1=0$

By definition, $e= (1+\frac{1}{n})^n, n \to \infty$

$e^z= (1+\frac{1}{n})^{nz}$

Let $N =nz$, $\frac{1}{n} = \frac{z}{N}$, then, $e^z= (1+\frac{z}{N})^{N}$

Let $z = i\pi$, then $e^{i\pi}= (1+\frac{i\pi}{N})^{N}$

When $N\to \infty$, $(1+\frac{i\pi}{N})^{N} \to -1$

So,  $ e^{i\pi}+1=0$, This is the **Euler's equation**. It contains $0,1,e,\pi,i$. The independently developed concepts finally come together to one equation.

<p><a href="https://commons.wikimedia.org/wiki/File:ExpIPi.gif#/media/File:ExpIPi.gif"><img  width="300" height="300" src="https://upload.wikimedia.org/wikipedia/commons/0/0e/ExpIPi.gif" alt="ExpIPi.gif"></a><br>By <a href="//commons.wikimedia.org/wiki/User:Sbyrnes321" title="User:Sbyrnes321">Sbyrnes321</a> - <span class="int-own-work" lang="en">Own work</span>, Public Domain, <a href="https://commons.wikimedia.org/w/index.php?curid=3997246">Link</a></p>

**Euler's identity** is a special case of Euler's formula from complex analysis, which states that for any real number x,

$${\displaystyle e^{ix}=\cos x+i\sin x}$$

where the inputs of the trigonometric functions sine and cosine are given in radians.
In particular, when x = π, or one half-turn (180°) around a circle:

$${\displaystyle e^{i\pi }=\cos \pi +i\sin \pi .}$$

Since $\cos \pi =-1$, and $\sin \pi =0,$ it follows that ${\displaystyle e^{i\pi }=-1+0i,}$ which yields Euler's identity: 

$${\displaystyle e^{i\pi }+1=0.}$$

```python
################################################################################
# Catenary and Parabola
# Author: stevenlee, tw.steven@gmail.com
# Date: 2017.05.19
################################################################################
import numpy as np
import matplotlib.pyplot as plt

a=5.
x = np.arange(-10., 10.1, 0.1)
y_catenary = a*(np.exp(x/a)+np.exp(-x/a))*0.5

c = 0.13
d = 5.
y_parabloa = c*(x**2) + d

plt.figure(figsize=(6, 6), dpi=80)
ax = plt.gca()
ax.cla()

plt.plot(x,y_catenary,'g')
plt.plot(x,y_parabloa,'r')

plt.grid(True)
plt.title('Catenary (green) and Parabola (red)')
plt.ylabel('y')
plt.xlabel('x')
plt.axis('equal')
plt.show()
```
